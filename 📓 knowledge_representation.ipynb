```python
facts = [
    ("Uhuru", "is_a", "human"),
    ("Raila", "is_a", "human"),
    ("human", "is_mortal", "true")
]

print("Initial Facts:")
for fact in facts:
    print(fact)
```

```python
inferred_facts = []

for fact in facts:
    subject, relation, obj = fact
    if relation == "is_a" and obj == "human":
        inferred_facts.append((subject, "is_mortal", "true"))

print("\nInferred Facts:")
for fact in inferred_facts:
    print(fact)
```

```python
# Combine original and inferred facts
all_facts = facts + inferred_facts

# Query: Who is mortal?
query_results = []

for fact in all_facts:
    subject, relation, obj = fact
    if relation == "is_mortal" and obj == "true":
        query_results.append(subject)

print("\nQuery Result:")
print("Mortal beings:", query_results)

```
